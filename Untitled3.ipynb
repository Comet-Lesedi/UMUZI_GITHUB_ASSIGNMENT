```python
# Load the dataset
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Load and explore the data
df = pd.read_csv('Wholesale customers data.csv')
print("First 5 rows of the dataset:")
print(df.head())

# Step 2: Prepare the data (pick numeric columns and drop missing)
cleaneddf = df.select_dtypes(include='number').dropna()
print("\nCleaned DataFrame shape:", cleaneddf.shape)

# Step 3: Standardize the data
scaler = StandardScaler()
sd = scaler.fit_transform(cleaneddf)

# Step 4: Apply PCA (reduce to 2 components for visualization)
pca = PCA(n_components=2)
rpca = pca.fit_transform(sd)
print("\nExplained variance ratio:", pca.explained_variance_ratio_)

# Step 5: Visualize PCA result
plt.figure(figsize=(8, 6))
plt.scatter(rpca[:, 0], rpca[:, 1], color='skyblue')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of the Wholesale Customer Dataset")
plt.grid(True)
plt.show()

# Step 6: Determine optimal number of clusters using Elbow Method
inertia = []
rk = range(1, 11)

for k in rk:
    kme = KMeans(n_clusters=k, random_state=42, n_init=10)
    kme.fit(sd)
    inertia.append(kme.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(rk, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()

# Step 7: Apply KMeans clustering with chosen k (e.g., k=3)
kopt = 3
kme = KMeans(n_clusters=kopt, random_state=42, n_init=10)
clu = kme.fit_predict(sd)

# Step 8: Add cluster labels to cleaned dataframe
cleaneddf['Cluster'] = clu

# Step 9: Visualize clusters in PCA space
plt.figure(figsize=(8, 6))
for cluster_id in range(kopt):
    plt.scatter(
        rpca[clu == cluster_id, 0],
        rpca[clu == cluster_id, 1],
        label=f'Cluster {cluster_id}'
    )

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Customer Segments via KMeans Clustering")
plt.legend()
plt.grid(True)
plt.show()

# Step 10: Evaluate clustering using silhouette score
sil_score = silhouette_score(sd, clu)
print(f"\nSilhouette Score: {sil_score:.3f}")

# Step 11: Interpret clusters
cluster_profiles = cleaneddf.groupby('Cluster').mean()
print("\nCluster Profiles (average feature values):")
print(cluster_profiles)
```
